```python
# select the relevant feature for my input values(x)
x=df[['cylinders','displacement','horsepower','weight']]# you can select other feature here
y=df['mpg']#target variable
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=20)
```

```python
x_test.shape
```

```python
x_train.shape
```

```python
y_test.shape
```

```python
y_train.shape
```

```python
# create polyn0minal features
degree= 2 #change the degree of the polnominal
poly=PolynomialFeatures(degree)
x_poly_train=poly.fit_transform(x_train)
```

```python
# fit a polynomal regression model
model=LinearRegression()
model.fit(x_poly_train,y_train)
```

```python
# make predictions of y
# polynomially transform the X_test values for prediction
x_poly_test= poly.transform(x_test)
y_pred= model.predict(x_poly_test)
```

```python
# Evaluate the model on the test set 
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
r2 = r2_score(y_test,y_pred)
print(f'mean squared Error: {mse:.2f}')
print(f' Root mean squared Error(RMSE):{rmse:.2f}')
print(f' R-squared(R^2) or Accuracy :{r2*100:.2f}%')

```
